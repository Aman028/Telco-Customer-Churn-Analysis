```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Replacing blanks with 0 and converted totalcharges into float datatype**

```python
df['TotalCharges']=df['TotalCharges'].replace(' ','0')
```

```python
df['TotalCharges']=df['TotalCharges'].astype('float')
```

# converted 0 and 1 value of senior citizen to yes/no to make it easier


```python
def conv(value):
  if value==1:
    return "yes"
  else:
    return "no"

df['SeniorCitizen']=df['SeniorCitizen'].apply(conv)
```

```python
df.head()
```

```python
ax=sns.countplot(x=df['Churn'], data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customer by Churn")
plt.show()
```

```python
plt.figure(figsize=(3,4))
gb=df.groupby('Churn').agg({'Churn':'count'})
plt.pie(gb['Churn'],labels=gb.index,autopct = "%1.2f%%")
plt.title("percentage of Churned Customer", fontsize=10)
plt.show()
```

# from the given pie chart we can conclude that 26.54% of our customers have churned out.
**now let's explore reason behind that**


```python
plt.figure(figsize=(3,3))
sns.countplot(x="gender", data=df,hue="Churn")
plt.title("Churn by Gender")
plt.show
```

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(x="SeniorCitizen", data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of customer by SeniorCitizen")
plt.show
```

```python
count_data = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)


percent_data = count_data.div(count_data.sum(axis=1), axis=0) * 100

ax = percent_data.plot(kind='bar', stacked=True, figsize=(4, 4))
plt.title("Churn by Senior Citizen (Percentage)")
plt.ylabel("Percentage of Total")
plt.xlabel("Senior Citizen")
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'], rotation=0)
plt.legend(title='Churn',bbox_to_anchor=(0.9,0.9))

for i in range(len(percent_data)):
    for j in range(len(percent_data.columns)):
        value = percent_data.iloc[i, j]
        ax.text(i, percent_data.iloc[i, :j + 1].sum() - value / 2, f"{value:.1f}%",
                ha='center', va='center', color='black')

plt.tight_layout()
plt.show()
```

**comparative a greater percentage of people in senior citizen category have churned out**

```python
plt.figure(figsize=(9,4))
sns.histplot(x="tenure", data=df,bins=72, hue="Churn")
plt.title("Churn by Tenure")
plt.show()
```

people who have used our services for a long time have stayed for a long time and people who have used our services 1 or 2 months have churn

```python
plt.figure(figsize=(4,4))
ax=sns.countplot(x="Contract", data=df, hue="Churn")
ax.bar_label(ax.containers[0])
plt.title("Count of customer by Contract")
plt.show
```

People who have month to month contract are likely to churn then from those who have 1 or 2 years of contract

```python
df.columns.values
```

```python

# List of columns for which you want to create count plots
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set the number of subplots
num_columns = len(columns)
fig, axes = plt.subplots(nrows=(num_columns + 2) // 3, ncols=3, figsize=(15, 5 * ((num_columns + 2) // 3)))
axes = axes.flatten()  # Flatten the 2D array of axes

# Create a count plot for each column
for i, column in enumerate(columns):
    sns.countplot(data=df, x=column, ax=axes[i], hue=df["Churn"])
    axes[i].set_title(f'Count Plot of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Count')

# Remove any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

- **Customers with phone service and multiple lines show a higher tendency to churn compared to those without.**
- **Fiber optic internet service customers exhibit higher churn rates than DSL users.**
- **Lack of services like online security, backup, and tech support correlates with higher churn rates, while the presence of these services tends to retain customers.**
